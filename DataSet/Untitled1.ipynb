```python
from ultralytics import YOLO

# Buat model dari YOLOv8 versi kecil (kamu bisa ganti ke yolov8s.pt untuk lebih akurat)
model = YOLO('yolov8n.pt')

# Jalankan training
model.train(
    data="/content/Herbal_Detection_Dataset/data.yaml",  # path dataset
    epochs=50,      # jumlah iterasi training
    imgsz=640,      # ukuran gambar input
    batch=16,       # ukuran batch
    project="/content/runs",
    name="train_herbal"
)
```
